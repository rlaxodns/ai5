# <font color=blue> Bag of Words Representation </font>

## BOW Representation using Scikit-learn's CountVectorizer
Text preprocessing, tokenizing and the ability to filter out stopwords are all included in [CountVectorizer](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html), which builds a dictionary of features and transforms documents to feature vectors.

## Scikit-learn's CountVectorizer Methods

<table>
<tr><td>`fit(raw_documents[, y])`</td><td>Learn a vocabulary dictionary of all tokens in the raw documents.</td></tr>
<tr><td>`transform(raw_documents)`</td><td>Transform documents to document-term matrix.</td></tr>
<tr><td>`fit_transform(raw_documents[, y])`</td><td>Learn the vocabulary dictionary and return document-term matrix.</td></tr>
<tr><td>`get_feature_names_out([input_features])`</td><td>Get output feature names for transformation.</td></tr>
<tr><td>`vocabulary_`</td><td>A dictionary where keys are terms and values are indices in the feature matrix.</td></tr>

</table>

```python
import spacy

nlp = spacy.load('en_core_web_sm')
```

```python
corpus = [
    'i like this film',
    'the film parasite is an excellent film',
    'bong is an excellent director',
    'i would give this film five out of five stars'
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
# Transform the documents into a bag-of-words representation
corpus_cv = cv.fit_transform(corpus)
corpus_cv
```

```python
print(corpus_cv.toarray())
```

```python
cv.get_feature_names_out()
```

```python
# Convert a bag-of-words representation to an array
import pandas as pd

df = pd.DataFrame(corpus_cv.toarray(), columns = cv.get_feature_names_out())
print(df)
```

## Let's Revisit the Movie Review Example
>- <font color = green> preprocess </font>
>- <font color = green> remove 'five' from the stop word list </font>

```python
# Remove the word from the set of stop words
nlp.Defaults.stop_words.remove('five')

nlp.vocab['five'].is_stop = False
```

```python
nlp.vocab['five'].is_stop
```

```python
# Preprocess

def preprocess(doc):
    doc = nlp(doc)

    preprocessed_tokens = []
    for token in doc:
        if token.is_punct or token.is_stop:
            continue
        else:
            preprocessed_tokens.append(token.lemma_)
    return " ".join(preprocessed_tokens)
```

```python
preprocessed_corpus = [preprocess(d) for d in corpus]
print(preprocessed_corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
corpus_cv = cv.fit_transform(preprocessed_corpus)
corpus_cv
```

```python
corpus_cv.toarray()
```

```python
cv.get_feature_names_out()
```

```python
import pandas as pd
bow_df = pd.DataFrame(corpus_cv.toarray(), columns = cv.get_feature_names_out())
print(bow_df)
```

# <font color=blue> Stop Words </font>

Words like "a" and "the" appear so frequently that they don't require tagging as thoroughly as nouns, verbs and modifiers. We call these *stop words*, and they can be filtered from the text to be processed. spaCy holds a built-in list of some 305 English stop words.

```python
import spacy

nlp = spacy.load('en_core_web_sm')
```

```python
# Print the set of spaCy's default stop words (remember that sets are unordered):
print(nlp.Defaults.stop_words)
```

```python
len(nlp.Defaults.stop_words)
```

## To see if a word is a stop word

```python
nlp.vocab['is'].is_stop
```

```python
nlp.vocab['covid'].is_stop
```

## To add a stop word
There may be times when you wish to add a stop word to the default set. Perhaps you decide that `'btw'` (common shorthand for "by the way") should be considered a stop word.

```python
# Add the word to the set of stop words. Use lowercase!
nlp.Defaults.stop_words.add('btw')

nlp.vocab['btw'].is_stop = True
```

```python
len(nlp.Defaults.stop_words)
```

```python
nlp.vocab['btw'].is_stop
```

## To remove a stop word

You may decide that `'beyond'` should not be considered a stop word.

```python
# Remove the word from the set of stop words
nlp.Defaults.stop_words.remove('beyond')

nlp.vocab['beyond'].is_stop = False
```

```python
len(nlp.Defaults.stop_words)
```

```python
nlp.vocab['beyond'].is_stop
```

# <font color=blue> Running Example: Trump Tweets </font>

- Pre-processing the text
- BOW Representation

<font color=green> Step 1. Pre-processing </font>

```python
tweets_trump = [
    'First Vaccine Administered. Congratulations USA! Congratulations WORLD!',
    'People working in the White House should receive the vaccine somewhat later in the program, unless specifically necessary. I have asked that this adjustment be made. I am not scheduled to take the vaccine, but look forward to doing so at the appropriate time. Thank you!',
    'Until Noverber 4th., Fake News Media is going full on Covid, Covid, Covid. We are rounding the turn, 99.9%'
]
```

```python
# Get Lowercase Tweet
tweets_trump = [tweet.lower() for tweet in tweets_trump]
print(tweets_trump)
```

```python
def preprocess(doc):
    doc = nlp(doc)

    preprocessed_tokens = []
    for token in doc:
        if token.is_punct or token.is_stop:
            continue
        else:
            preprocessed_tokens.append(token.lemma_)
    return " ".join(preprocessed_tokens)
```

```python
# Apply preprocess to each tweet
preprocessed_tweets = [preprocess(t) for t in tweets_trump]
print(preprocessed_tweets)
```

<font color=green> Step 2. BOW Representation </font>

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
# Transform the documents into a bag-of-words representation
tweets_cv = cv.fit_transform(preprocessed_tweets)
tweets_cv
```

```python
print(tweets_cv.toarray())
```

```python
print(cv.get_feature_names_out())
```

```python
# Create BOW Representation Dataframe
import pandas as pd
df = pd.DataFrame(tweets_cv.toarray(), columns = cv.get_feature_names_out())
print(df)
```
