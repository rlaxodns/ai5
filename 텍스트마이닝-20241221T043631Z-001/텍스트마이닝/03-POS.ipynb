## 1. View token tags

* To view the coarse POS tag use `token.pos_`
* To view the fine-grained tag use `token.tag_`
* To view the description of either type of tag use `spacy.explain(tag)`

<div class="alert alert-success">Note that `token.pos` and `token.tag` return integer hash values; by adding the underscores we get the text equivalent that lives in **doc.vocab**.</div>

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("The quick brown fox jumped over the lazy dog's back.")
```

```python
# Get 'jumped'
doc[4]
```

```python
# view token tags
print(doc[4].text, doc[4].pos_, doc[4].tag_, spacy.explain(doc[4].tag_))
```

**Apply this to the entire Doc object:**

```python
for token in doc:
    print(token.text, token.pos_, token.tag_, spacy.explain(token.tag_))
```

**Formatting**

```python
for token in doc:
    print(f'{token.text:{10}} {token.pos_:{10}} {token.tag_:{10}} {spacy.explain(token.tag_)}')
```

## <font color = blue> Different POS tags *depending on the context* </font>

In the first example, with no other cues to work from, spaCy assumed that ***read*** was present tense, while in the second example with ***the***, spaCy assigned the past tense.

```python
doc1 = nlp("I read a book on NLP")
```

```python
word1 = doc1[1]
word1.text
```

```python
print(f'{word1.text:{10}} {word1.pos_:{10}} {word1.tag_:{10}} {spacy.explain(word1.tag_)}')
```

```python
doc2 = nlp("I read the book on NLP")
```

```python
word2 = doc2[1]
word2.text
```

```python
print(f'{word2.text:{10}} {word2.pos_:{10}} {word2.tag_:{10}} {spacy.explain(word2.tag_)}')
```

## 2. Counting POS Tags
The `Doc.count_by()` method accepts a specific token attribute as its argument, and returns a frequency count of the given attribute as a dictionary object.

*  **Keys**: integer values of the given attribute ID (e.g., POS tags)
*  **Values**: frequency of the given attribute

NOTE: counts of zero are not included.

```python
doc = nlp("The quick brown fox jumped over the lazy dog's back.")
```

```python
# Count the frequencies of different coarse-grained POS tags:
POS_counts = doc.count_by(spacy.attrs.POS)
POS_counts
```

### Decoding POS tag ID

```python
print(doc[2].text)
print(doc[2].pos)
print(doc[2].pos_)
```

```python
doc.vocab[84].text
```

### Create a frequency list of POS tags (corase-grained) from the entire document

```python
# List of tuples
POS_counts.items()
```

```python
# k indicates POS Id
# v indicates frequency

for k,v in POS_counts.items():
    print(f'{k:<{5}} {doc.vocab[k].text:{8}} {v}')
```

```python
# Sorted
for k,v in sorted(POS_counts.items()):
    print(f'{k:<{5}} {doc.vocab[k].text:{8}} {v}')
```

## 3. Visualizing Parts of Speech
spaCy offers an outstanding visualizer called **displaCy**:

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# Import displaCy
from spacy import displacy
```

```python
doc = nlp("The quick brown fox jumped over the lazy dog's back.")
```

```python
# Render the dependency parse immediately inside Jupyter:
displacy.render(doc, style='dep', jupyter=True, options={'distance': 110})
```
