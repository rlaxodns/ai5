# 1. Basics of Regular Expressions

Regular Expression (also called RE or RegEx, in short) is a sequence of characters that forms a search pattern. It is used to check whether the search pattern exists in the given string.

Regular expressions are handled using Python's built-in **re** library. See [the docs](https://docs.python.org/3/library/re.html) for more information.

## Searching for Basic Patterns

```python
text = "My phone number is 765-464-9593. Call me!"
```

```python
'765-464-9593' in text
```

The format of regular expressions

```python
# re for Regular Expression
import re
```

```python
# Define pattern
pattern = 'phone'
```

```python
re.search(pattern, text)
```

<font color=green> Methods of regular expressions </font>
- `.search()`
- `.findall()`

But what if the pattern occurs more than once?

```python
text2 = "My phone is a new phone"
pattern = "phone"
```

```python
matches = re.search(pattern, text2)
matches
```

```python
matches = re.findall(pattern, text2)
matches
```

# 2. Patterns

Regular expressions to find out a pattern of text strings <br>
For example,
- uppercase and lowercase of a certain word in a large string of text(e.g., woodchuck/ Woodchuck)
- the general format of a telephone number or an email address

```python
pattern = "\d\d\d-\d\d\d-\d\d\d\d"

matches = re.findall(pattern, text)
matches
```

## Quantifiers

<br>
<table ><tr><th>Character</th><th>Description</th><th>Example Pattern Code</th><th >Exammple Match</th></tr>

<tr ><td><span >+</span></td><td>Occurs one or more times</td><td>	Version \w-\w+</td><td>Version A-b1_1</td></tr>

<tr ><td><span >{3}</span></td><td>Occurs exactly 3 times</td><td>\D{3}</td><td>abc</td></tr>



<tr ><td><span >{2,4}</span></td><td>Occurs 2 to 4 times</td><td>\d{2,4}</td><td>123</td></tr>



<tr ><td><span >{3,}</span></td><td>Occurs 3 or more</td><td>\w{3,}</td><td>anycharacters</td></tr>

<tr ><td><span >\*</span></td><td>Occurs zero or more times</td><td>A\*B\*C*</td><td>AAACC</td></tr>

</table>

```python
text = "My phone number is 765-464-9593. Call me!"
```

```python
#pattern = "\d\d\d-\d\d\d-\d\d\d\d"
pattern = "\d{3}-\d{3}-\d{4}"

matches = re.findall(pattern, text)
matches
```

- Letters inside square brackets [ ]

```python
text3 = """Elephant comprises various species,
including the African forest elephant and the Asian elephant
"""
```

```python
# lowercase e and Uppercase E
matches = re.findall("[eE]lephant", text3)
matches
```

- Ranges [ ]

```python
text4 = "I ate 99 slices of pizza and 108 cookies"
```

```python
matches = re.findall("[0-9]+", text4)
matches
```

## Additional Regex Syntax

### Or operator |

```python
text5 = "woodchuck is also called as groundhog"
```

```python
matches = re.findall("woodchuck|groundhog", text5)
matches
```

### Negation (Exclusion)

- `[^ ]`: Except anything inside the brackets

```python
text6 = "there are 3 numbers 34 inside 5 this sentence."
```

```python
matches = re.findall("[^\d+]+", text6)
matches
```

remove punctuation

```python
text7 = "This is a string! But it has punctuation. How can we remove it?"
```

```python
matches = re.findall("[^!.?]+", text7)
matches
```

### The Wildcard Character

- ` . `: Use a "wildcard" as a placement that will match any character
placed there.
- ` * `: 0 or more times
- ` + `: 1 or more times

```python
text8 = """She decided to being her journey,
and she began to realize that the journey had just begun.
"""
```

```python
matches = re.findall("beg.n", text8)
matches
```

with ` * `, see what happens

```python
text9 = "Look at the cat in the hat that sat splat"
```

```python
matches = re.findall(".at", text9)
matches
```

With ` * `, see what happens

```python
matches = re.findall("\S*at", text9)
matches
```

With ` + `, see what happens

```python
matches = re.findall("\S+at", text9)
matches
```

# <font color = blue> Text Cleaning with Substitute </font>

`re.sub(pattern, repl, string)`

```python
# Remove punctuation
text6 = "there are 3 numbers 34 inside 5 this sentence."
text7 = "This is a string! But it has punctuation. How can we remove it?"

text6 = re.sub("[^a-zA-Z\s]", "", text6)
print(text6)

text7 = re.sub("[^a-zA-Z\s]", "", text7)
print(text7)
```

```python
text8 = "This is 3 string! But it 45 has punctuations. How 7 can we remove it?"

text8_a = re.sub("\d+", "", text8)
text8_b = re.sub("[!.?]", "", text8_a)
print(text8_b)
```
