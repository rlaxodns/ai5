# <font color=blue>1. spaCy</font>

```python
# Import spaCy and load the language library
import spacy
nlp = spacy.load('en_core_web_sm')
```

<img src="tokenization.png" width="550">

-  **Prefix**:	Character(s) at the beginning &#9656; `$ ( “ ¿`
-  **Suffix**:	Character(s) at the end &#9656; `km ) , . ! ”`
-  **Infix**:	Character(s) in between &#9656; `- -- / ...`
-  **Exception**: Special-case rule to split a string into several tokens or prevent a token from being split when punctuation rules are applied &#9656; `St. U.S.`

```python
string1 = '"we\'re moving to L.A.!"'
```

```python
print(string1)
```

```python
doc1 = nlp(string1)
```

```python
for token in doc1:
    print(token.text)
```

<font color=green>`spaCy` will isolate punctuation that does *not* form an integral part of a word. Quotation marks, commas, and punctuation at the end of a sentence will be assigned their own token. </font>
> <font color=green> However, punctuation that exists as part of an email address, website or numerical value will be kept as part of the token.</font>

```python
string2 = "AT&T offers a $49.99 montly plan for N.Y. citizens who enjoy rock'n'roll."
```

```python
doc2 = nlp(string2)
for token in doc2:
    print(token.text)
```

```python
string3 = "If you'd like to join Ph.D. program, email us at userid@dongguk.edu or visit at https://www.dongguk.edu"
```

```python
doc3 = nlp(string3)
for token in doc3:
    print(token.text)
```

## 2.2. WordPunctTokenizer

```python
from nltk.tokenize import WordPunctTokenizer
```

- string3 = "If you'd like to join Ph.D. program, email us at userid@dongguk.edu or visit at https://www.dongguk.edu"

## 2.3. TweetTokenizer

For twitter data we cannot apply the spacy trained model but instead we use TweetTokenizer from NLTK library. This will separate # tags and @ tags and symbols as a separate token.

```python
tweet = "Snow White and the Seven Degrees #MakeAMovieCold@midnight :-)"
```

- spaCy

```python
doc = nlp(tweet)
for token in doc:
    print(token.text, end=' | ')
```

- TweetTokenizer

```python
from nltk.tokenize import TweetTokenizer

tokenizer = TweetTokenizer()
print(tokenizer.tokenize(tweet))
```

# <font color=blue> 3. Tokenization using Regex </font>

```python
import re

text = "The quick brown   fox jumps over the    lazy dog"
```

```python
re.split("\s+", text)
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
text = "AT&T offers a $49.99 montly plan for N.Y. citizens"
```

```python
tokenizer1 = RegexpTokenizer("\w+")
tokenizer2 = RegexpTokenizer("\s+", gaps=True)

print(tokenizer1.tokenize(text))
print(tokenizer2.tokenize(text))
```
