## 1.1. Porter Stemmer

One of the most common - and effective - stemming tools is [*Porter's Algorithm*](https://tartarus.org/martin/PorterStemmer/) developed by Martin Porter in [1980](https://tartarus.org/martin/PorterStemmer/def.txt). The algorithm employs five phases of word reduction, each with its own set of mapping rules. In the first phase, simple suffix mapping rules are defined, such as:

<img src="stemming1.png" width="450">

From a given set of stemming rules only one rule is applied, based on the longest suffix S1. Thus, `caresses` reduces to `caress` but not `cares`.

More sophisticated phases consider the length/complexity of the word before applying a rule. For example:

<img src="stemming2.png" width="600">

```python
import nltk

from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
```

```python
words = ['runs', 'ran', 'running', 'runner', 'goes', 'went', 'mice']
```

```python
for word in words:
    print(word + ' ---> ' + porter_stemmer.stem(word))
```

```python
words2 = ['computer', 'organization', 'easily', 'fairly', 'friendly']
```

```python
for word in words2:
    print(word + ' ---> ' + porter_stemmer.stem(word))
```

```python
words3 = ['generous', 'generously', 'generate', 'generation']
```

```python
for word in words3:
    print(word + ' ---> ' + porter_stemmer.stem(word))
```

## 1.2. Snowball Stemmer

Snowball is the name of a stemming language developed by Martin Porter. The algorithm used here is more acurately called the "English Stemmer" or "Porter2 Stemmer". It offers a slight improvement over the original Porter stemmer, both in logic and speed. Since **nltk** uses the name SnowballStemmer, we'll use it here.

```python
from nltk.stem.snowball import SnowballStemmer

sb_stemmer = SnowballStemmer(language='english')
```

```python
words2 = ['computer', 'organization', 'easily', 'fairly', 'friendly']
```

```python
for word in words2:
    print(word + ' ---> ' + sb_stemmer.stem(word))
```

```python
words3 = ['generous', 'generously', 'generate', 'generation']
```

```python
for word in words3:
    print(word + ' ---> ' + sb_stemmer.stem(word))
```

# <font color=blue>2. Lemmatization</font>

```python
import spacy

nlp = spacy.load('en_core_web_sm')
```

```python
words = 'runs ran running runner goes went mice'
```

```python
doc = nlp(words)

for token in doc:
    print(token.text, token.lemma_,  token.lemma,  token.pos_)
```

```python
for token in doc:
    print(f'{token.text:{10}} {token.lemma_:{10}} {token.lemma:<{25}} {token.pos_:{10}}')
```

<font color=green>Notice that `runs`, `ran` and `running` all point to the same lemma `run` (12767...).</font>

```python
# Function to display lemmas
def show_lemmas(doc):
    for token in doc:
        print(f'{token.text:{15}} {token.pos_:{10}} {token.lemma:<{25}} {token.lemma_:{10}}')
```

<font color = green> Recall </font>
- words2 = 'computer organization easily fairly friendly'
- words3 = 'generous generously generate generation'

```python
words2 = 'computer organization easily fairly friendly'
words2_doc = nlp(words2)

show_lemmas(words2_doc)
```

```python
words3 = 'generous generously generate generation'
words3_doc = nlp(words3)

show_lemmas(words3_doc)
```

<font color = green> Let's revisit the phrase `I am meeting him tomorrow at the meeting` </font>

```python
phrase = 'I am meeting him tomorrow at the meeting'
doc = nlp(phrase)
```

```python
show_lemmas(doc)
```

<font color=green>Here the lemma of `meeting` is determined by its Part of Speech tag.</font>

# <font color=blue> Exercise. </font>

<b> ex 1. </b> love, loves, loved, loving <br>
<b> ex 2. </b> innovation, innovations, innovate, innovates, innovative

```python
# Porter's stemmer
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()

words1 = ['love', 'loves', 'loved', 'loving']
for word in words1:
    print(word + ' ---> ' + porter_stemmer.stem(word))
```

```python
words2 = ['innovation', 'innovations', 'innovate', 'innovates', 'innovative']
for word in words2:
    print(word + ' ---> ' + porter_stemmer.stem(word))
```

```python
# Lemmatization
import spacy
nlp = spacy.load('en_core_web_sm')

words1 = 'love loves loved loving'
words1 = nlp(words1)

for token in words1:
    print(f'{token.text:{10}} {token.lemma_:{10}} {token.lemma:<{25}} {token.pos_:{10}}')
```

```python
words2 = 'innovation innovations innovate innovates innvative'
words2 = nlp(words2)

for token in words2:
    print(f'{token.text:{15}} {token.lemma_:{15}} {token.lemma:<{25}} {token.pos_:{10}}')
```
